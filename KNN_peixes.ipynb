```python
#Ordinal encoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
ordinal = OrdinalEncoder()
```

```python
species_cat = dataframe['Species']
```

```python
#Transformando as especies (dado categorico) em numeros
species_num = ordinal.fit_transform(species_cat.values.reshape(-1, 1))
```

```python
#Variavel que armazena as categorias
ordinal.categories_[0]
```

```python
#Criando um dic com as categocias e seus respectivos valores
encoding_dict = dict(zip((ordinal.categories_[0]), range (len(ordinal.categories_[0]))))
```

```python
encoding_dict
```

```python
#Substituindo os valores categorico pelos dados convertidos
dataframe['Species'] = species_num
```

```python
dataframe
```

```python
#Pegando as variaveis independentes x (tudo menos o target -> Species)
x = dataframe.drop(['Species'], axis=1)
```

```python
x
```

```python
y = dataframe['Species']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
#Modelo de Regressao
model_knn = KNeighborsClassifier(n_neighbors=3)
```

```python
#Treinando o modelo
model_knn.fit(X_train, y_train)
```

```python
y_predict = model_knn.predict(X_test)
```

```python
y_predict
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_predict)
```

```python
list_result = list()
for n_knn in range(1,100,2):
  model_knn = KNeighborsClassifier(n_neighbors=n_knn)
  model_knn.fit(X_train, y_train)
  y_predict = model_knn.predict(X_test)
  acc = accuracy_score(y_test, y_predict)
  list_result.append([acc, n_knn])
```

```python
#Criando um dataframe dos resultados
dataframe_result = pd.DataFrame(list_result, columns=['Acurácia','N-Vizinhos'])
```
